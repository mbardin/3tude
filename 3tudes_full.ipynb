```python
 # "H.K": An audification of COVID data from Hong Kong from 01-22-20 to 09-10-21
    
dur = 120 #seconds
durInSamples = fs * dur
completeAudio = np.zeros((2,durInSamples))
    
with open('hong_kong_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        HKCases = np.array(next(reader))
    
HKCases = HKCases.astype(float)
    
     # total_duration is in samples
    # s is the relative height of the sustain portion
    # A,D,S,R describe the timing of the ADSR envelope components
def ADSRenvelope(total_duration, s, A, D, S, R):
    normalize_it = A+D+S+R
    
    ADpart = np.linspace(0, 1, round(float(A)/float(normalize_it)*total_duration))
    DSpart = np.linspace(1, s, round(float(D)/float(normalize_it)*total_duration))
    SRpart = np.linspace(s, s, round(float(S)/float(normalize_it)*total_duration))
    Rpart = np.linspace(s, 0, round(float(R)/float(normalize_it)*total_duration))
    
    ADSR = np.concatenate((ADpart, DSpart, SRpart, Rpart))
    
    if (np.size(ADSR) - total_duration) > 0:
        ADSR = ADSR[1:total_duration]
        print ('ADSR too long')
    
    if (np.size(ADSR) - total_duration) < 0:
        ADSR = np.concatenate((ADSR, np.zeros(total_duration-np.size(ADSR))))
        print ('ADSR too short')

    return ADSR
    
def makesequence(values):
    seq = np.empty(0)
    
    note = generatesinewave(values, dur)
    gain = 0.3
    note = note * gain
    seq = np.concatenate((seq, note))
    return seq

def overlap(m,datatoinsert,therow,thecol):
    len = np.size(datatoinsert)
    priorData = m[therow,thecol:(thecol+len)]
    priorData = priorData+datatoinsert.T
    m[therow,thecol:(thecol+len)] = priorData
    
    # panval should be between 0 and 1
def overlapStereo(m,datatoinsert,panval,thecol):
    if panval > 1:
        panval = 1
    if panval < 0:
        panval = 0
    
    panVal2 = 1-panval
    panval = np.sqrt(panval)
    panVal2 = np.sqrt(panVal2)
    
    overlap(m,panval*datatoinsert,0,thecol)
    overlap(m,panVal2*datatoinsert,1,thecol)
    
    
    
    
for n in HKCases:
    t = np.linspace(float(0), dur, int(durInSamples))
    note = makesequence(n)
    #frequency = mtof(note)
    #mysinewave = np.sin(2*np.pi*t*frequency)
    #mysinewave = mysinewave * (t / dur) * (t / dur) * (t / dur)
    envelope = ADSRenvelope(np.size(note), durInSamples, 10, 10, 200, 20)
    note = note * envelope
    
    panval = n / max(HKCases)
    
    overlapStereo(completeAudio,note,panval,0)

    

# Make sure that the audio isn't clipping
completeAudio2 = completeAudio / np.max(np.max(np.abs(completeAudio)))

intSignal = np.int16(completeAudio2 * 32767)
scipy.io.wavfile.write('3tude_HK.wav', fs, intSignal.T)
    
```

```python

    # "4605457": A sonification of Covid Case and Death Data from
    # code takes in the csv, and uses the values of each column to make a MIDI note (1 per 500ms)
    # current version creates 8 audio files.and a mc version of those files
    # Code by Matthew A. Bardin [Dec 2021]
    
    # Alternations
    # etude by Matthew A. Bardin Dec [ 2021]

 
    
    # import csv files.
    with open('brazil_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        brazilCases = np.array(next(reader))
        brazilCases = brazilCases.astype(float)
    
    
    with open('egypt_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        egyptCases = np.array(next(reader))
    
    egyptCases = egyptCases.astype(float)
    
    
    with open('hong_kong_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        HKCases = np.array(next(reader))
    
    HKCases = HKCases.astype(float)
    
    with open('italy_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        italyCases = np.array(next(reader))
    
    italyCases = italyCases.astype(float)
    
    with open('japan_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        japanCases = np.array(next(reader))
    
    japanCases = japanCases.astype(float)
   
    with open('new_zealand_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        NZCases = np.array(next(reader))
    
    NZCases = NZCases.astype(float)
    
    with open('USA_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        USACases = np.array(next(reader))
    
    USACases = USACases.astype(float)
    
    with open('zimbabwe_cases.csv') as csv_object:
        reader = csv.reader(csv_object)
        zimbabweCases = np.array(next(reader))
    
    zimbabweCases = zimbabweCases.astype(float)

   
    # total_duration is in samples
    # s is the relative height of the sustain portion
    # A,D,S,R describe the timing of the ADSR envelope components
    def ADSRenvelope(total_duration, s, A, D, S, R):
        normalize_it = A+D+S+R
    
        ADpart = np.linspace(0, 1, round(float(A)/float(normalize_it)*total_duration))
        DSpart = np.linspace(1, s, round(float(D)/float(normalize_it)*total_duration))
        SRpart = np.linspace(s, s, round(float(S)/float(normalize_it)*total_duration))
        Rpart = np.linspace(s, 0, round(float(R)/float(normalize_it)*total_duration))
    
        ADSR = np.concatenate((ADpart, DSpart, SRpart, Rpart))
    
        if (np.size(ADSR) - total_duration) > 0:
            ADSR = ADSR[1:total_duration]
            print ('ADSR too long')
    
        if (np.size(ADSR) - total_duration) < 0:
            ADSR = np.concatenate((ADSR, np.zeros(total_duration-np.size(ADSR))))
            print ('ADSR too short')

        return ADSR
    
    # generates a sequence of notes up to 4 octaves, adjusted up by 30 MIDI notes to make audible
    
    def makesequences1(values):
        seq = np.empty(0)
        for x in values:
            pitch = (x % 48) + 30
            if pitch < 10:
                continue
    #             skips this loop and adds it to array if value is inaudible. extra room given for a more \"feeling\" effect on larger speakers (test. channels will need to be the same length.)
    #             could also set pitch to a specific value in order to make a drone effect. (plan b is current method doesn't work)
            note = generatesinewave(pitch, 0.5) # half-second duration
            gain = 0.3
            note = note * gain
            seq = np.concatenate((seq, note))
        return seq
    
    def makesequences2(values):
        seq = np.empty(0)
        for x in values:
            pitch = (x % 48) + 30
            if pitch < 10:
                continue
    #             skips this loop and adds it to array if value is inaudible. extra room given for a more \"feeling\" effect on larger speakers (test. channels will need to be the same length.)
    #             could also set pitch to a specific value in order to make a drone effect. (plan b is current method doesn't work)
            note = generatesquwave(pitch, 0.5) # half-second duration
            gain = 0.3
            note = note * gain
            seq = np.concatenate((seq, note))
        return seq
    
    def makesequences3(values):
        seq = np.empty(0)
        for x in values:
            pitch = (x % 48) + 30
            if pitch < 10:
                continue
    #             skips this loop and adds it to array if value is inaudible. extra room given for a more \"feeling\" effect on larger speakers (test. channels will need to be the same length.)
    #             could also set pitch to a specific value in order to make a drone effect. (plan b is current method doesn't work)
            note = generatesawwave(pitch, 0.5) # half-second duration
            gain = 0.3
            note = note * gain
            seq = np.concatenate((seq, note))
        return seq
    
    def makemysequence(values):
        seq = np.empty(0)
        for x in values:
            pitch = 69
            check = x
            note_length = np.random.randint(5) / 10
            if check % 2 == 1:
                note = generatesinewave(pitch, note_length)
            else:
                note = generatesquwave(pitch, note_length)
            gain = 0.5
            note = note * gain
            seq = np.concatenate((seq, note))
        return seq
    
    
    
    # generates a sinewave, squwave, and sawtooth of a given frequency and duration. based on function by E. Berdhal
    # m = midi note value
    def generatesinewave(m, duration):
        t = np.linspace(0, duration, int(duration*fs))
        frequency = mtof(m)
        mysinewave = np.sin(2*np.pi*t*frequency)
    
        envelope = ADSRenvelope(np.size(mysinewave), 0.3, 10, 10, 60, 20)
        mysinewave = mysinewave * envelope
        return mysinewave
    
    def generatesquwave(m, duration):
        t = np.linspace(0, duration, int(duration*fs))
        frequency = mtof(m)
        sig = np.sin(2 * np.pi * t)
        pwm = signal.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
    
        envelope = ADSRenvelope(np.size(pwm), 0.3, 10, 10, 60, 20)
        pwm = pwm * envelope
        return pwm

    def generatesawwave(m, duration):
        t = np.linspace(0, duration, int(duration*fs))
        frequency = mtof(m)
        mysaw = signal.sawtooth(2 * np.pi * 5 * t)
        
        envelope = ADSRenvelope(np.size(mysaw), 0.3, 10, 10, 60, 20)
        mysaw = mysaw * envelope
        return mysaw
    
    
    # converts data into a scaled audio wave sequence\n",
    # make certain voices have a different range/timbre?
    brazilSequence = np.empty(0)
    brazilSequence = makesequences1(brazilCases)
    egyptSequence = np.empty(0)
    egyptSequence = makesequences2(egyptCases)
    HKSequence = np.empty(0)
    HKSequence = makesequences3(HKCases)
    italySequence = np.empty(0)
    italySequence = makesequences1(italyCases)
    japanSequence = np.empty(0)
    japanSequence = makesequences2(japanCases)
    NZSequence = np.empty(0)
    NZSequence = makesequences3(NZCases)
    USASequence = np.empty(0)
    USASequence = makesequences1(USACases)
    zimbabweSequence = np.empty(0)
    zimbabweSequence = makesequences2(zimbabweCases)
    
    soloItaly = np.empty(0)
    soloItaly = makemysequence(italyCases)
    
    # exports 8 seperate sound file(s)                                                     
    
    i = 0
    while i < 8:
      if i == 0:
        intSignal = np.int16(brazilSequence * 32767)
        scipy.io.wavfile.write('brazil.wav', fs, intSignal)
      elif i == 1:
        intSignal = np.int16(egyptSequence * 32767)
        scipy.io.wavfile.write('egypt.wav', fs, intSignal)
      elif i == 2:    
        intSignal = np.int16(HKSequence * 32767)
        scipy.io.wavfile.write('hongkong.wav', fs, intSignal)
      elif i == 3:  
        intSignal = np.int16(italySequence * 32767)
        scipy.io.wavfile.write('italy.wav', fs, intSignal)
      elif i == 4: 
        intSignal = np.int16(japanSequence * 32767)
        scipy.io.wavfile.write('japan.wav', fs, intSignal)    
      elif i == 5:
        intSignal = np.int16(NZSequence * 32767)
        scipy.io.wavfile.write('newzealand.wav', fs, intSignal)
      elif i == 6:
        intSignal = np.int16(USASequence * 32767)
        scipy.io.wavfile.write('usa.wav', fs, intSignal) 
      elif i == 7:
        intSignal = np.int16(zimbabweSequence * 32767)
        scipy.io.wavfile.write('zimbabwe.wav', fs, intSignal)
    
      i += 1
    
    intSignal = np.int16(soloItaly * 32767)
    scipy.io.wavfile.write('3tude_alternations.wav', fs, intSignal)
```
